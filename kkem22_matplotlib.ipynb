```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('exam_data.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
year = data['year']
male = data['male']
female = data['female']
yearlytotal = data['total']
```

```python
plt.plot(year,yearlytotal)
plt.title('students taking XYZ exam')
plt.xlabel('year')
plt.ylabel('number of students')
plt.show()
```

```python
x_index = np.arange(len(year))

plt.plot(x_index,yearlytotal)
plt.xticks(ticks=x_index,labels=year)
plt.title('Students taking XYZ exam by year')
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.show()
```

```python
x_index = np.arange(len(year))

plt.plot(x_index,yearlytotal)
plt.plot(x_index,male)
plt.plot(x_index,female)
plt.xticks(ticks=x_index,labels=year)
plt.title('Students taking XYZ exam by year')
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.show()
```

```python
x_index = np.arange(len(year))

plt.plot(x_index,yearlytotal,label ='Total')
plt.plot(x_index,male, label = 'male')
plt.plot(x_index,female,label = 'Female')
plt.xticks(ticks=x_index,labels=year)
plt.legend()
plt.title('Students taking XYZ exam by year')
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.show()
```

```python
x_index = np.arange(len(year))

plt.bar(x_index,yearlytotal,label ='Total')
plt.bar(x_index,male, label = 'male')
plt.bar(x_index,female,label = 'Female')
plt.xticks(ticks=x_index,labels=year)
plt.legend()
plt.title('Students taking XYZ exam by year')
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.show()
```

```python
x_index = np.arange(len(year))
width = 0.2

plt.bar(x_index-width,yearlytotal,width=width,label ='Total')
plt.bar(x_index,male,width=width, label = 'male')
plt.bar(x_index+width,female,width=width,label = 'Female')
plt.xticks(ticks=x_index,labels=year)
plt.legend()
plt.title('Students taking XYZ exam by year')
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.show()
```
